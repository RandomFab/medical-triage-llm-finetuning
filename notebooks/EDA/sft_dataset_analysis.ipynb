# EXPLORATION

```python
import pandas as pd
from pathlib import Path

pd.set_option('display.max_colwidth', None)
ROOT_PROJECT = Path.cwd().parent.parent
print(ROOT_PROJECT)
```

```python
df = pd.read_parquet(ROOT_PROJECT / "data" / "processed" / "sft_dataset" / "sft_dataset_augmented.parquet")
print(df.dataset_name.value_counts())
df[df["dataset_name"] == 'ultramed'].head(30)
```

```python
total_count = df['question'].str.count('LOCATION').sum()
nb_words = df['question'].str.split().apply(len).sum()
print(f"Les balises apparaîssent {total_count} fois au total.")
print(f"Le nombre total de mots dans les questions est de {nb_words}.")
print(f"Les balises représentent {total_count/nb_words:.2%} des mots dans les questions.")
```

```python
total_count = df['answer'].str.count('<LOCATION>').sum()
nb_words = df['answer'].str.split().apply(len).sum()
print(f"Les balises apparaîssent {total_count} fois au total.")
print(f"Le nombre total de mots dans les réponses est de {nb_words}.")
print(f"Les balises représentent {total_count/nb_words:.2%} des mots dans les réponses.")
```

```python
# Nombre de lignes touchées (au moins 1 balise)
mask_answer = df['answer'].str.contains(r'<[^>]+>', regex=True)
mask_question = df['question'].str.contains(r'<[^>]+>', regex=True)
mask_any = mask_answer | mask_question

print(f"Lignes avec balises dans answer : {mask_answer.sum()} / {len(df)} ({mask_answer.mean():.1%})")
print(f"Lignes avec balises dans question : {mask_question.sum()} / {len(df)} ({mask_question.mean():.1%})")
print(f"Lignes touchées (question OU answer) : {mask_any.sum()} / {len(df)} ({mask_any.mean():.1%})")

print("\n--- Répartition par source ---")
print(df[mask_any].groupby('dataset_name').size())
print("\n--- Répartition par type de balise (answer) ---")
print(df['answer'].str.extractall(r'(<[^>]+>)')[0].value_counts())
```
